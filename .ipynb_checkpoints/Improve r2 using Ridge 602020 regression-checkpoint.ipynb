# Ridge regression with 60/20/20 time split

This notebook trains a Ridge regression using a time-aware split into:
- Training = first 60% (used to fit models),
- Validation = next 20% (used to pick hyperparameters),
- Test = last 20% (final evaluation).

The code is beginner-friendly and avoids data leakage: all imputers, encoders and scalers are fitted on the training set only and then applied to validation/test sets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# 1. Load and clean
df = pd.read_csv('walmart-sales-dataset-of-45stores.csv', parse_dates=['Date'], dayfirst=True)
df = df.dropna(subset=['Weekly_Sales']).copy()

# 2. Feature engineering
df = df.sort_values('Date').reset_index(drop=True)
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.dayofweek

group_cols = [c for c in ['Store','Dept'] if c in df.columns]
if group_cols:
    df = df.sort_values(group_cols + ['Date']).reset_index(drop=True)
    df['Sales_Lag1'] = df.groupby(group_cols)['Weekly_Sales'].shift(1)
    df['Sales_Roll4'] = df.groupby(group_cols)['Weekly_Sales'].transform(lambda x: x.shift(1).rolling(4).mean())
else:
    df['Sales_Lag1'] = df['Weekly_Sales'].shift(1)
    df['Sales_Roll4'] = df['Weekly_Sales'].shift(1).rolling(4).mean()

df = df.dropna().reset_index(drop=True)
df = df.drop(columns=['Date'])

# 3. Time-aware split
n = len(df)
cut1 = int(n * 0.6)
cut2 = int(n * 0.8)
train = df.iloc[:cut1].copy()
val = df.iloc[cut1:cut2].copy()
test = df.iloc[cut2:].copy()

# 4. Frequency encoding using TRAIN only
original_cat_cols = [c for c in ['Store','Dept'] if c in df.columns]
for c in original_cat_cols:
    freq_map = train[c].value_counts(normalize=True).to_dict()
    train[c + '_freq'] = train[c].map(freq_map).fillna(0)
    val[c + '_freq'] = val[c].map(freq_map).fillna(0)
    test[c + '_freq'] = test[c].map(freq_map).fillna(0)
    train.drop(columns=[c], inplace=True)
    val.drop(columns=[c], inplace=True)
    test.drop(columns=[c], inplace=True)

# 5. Define X/y
y_train = train['Weekly_Sales']
X_train = train.drop(columns=['Weekly_Sales'])
y_val = val['Weekly_Sales']
X_val = val.drop(columns=['Weekly_Sales'])
y_test = test['Weekly_Sales']
X_test = test.drop(columns=['Weekly_Sales'])

# 6. Impute and scale
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X_train.select_dtypes(include=['object','category']).columns.tolist()

num_imputer = SimpleImputer(strategy='mean')
X_train[num_cols] = num_imputer.fit_transform(X_train[num_cols])
X_val[num_cols] = num_imputer.transform(X_val[num_cols])
X_test[num_cols] = num_imputer.transform(X_test[num_cols])

if len(cat_cols) > 0:
    cat_imputer = SimpleImputer(strategy='most_frequent')
    X_train[cat_cols] = cat_imputer.fit_transform(X_train[cat_cols])
    X_val[cat_cols] = cat_imputer.transform(X_val[cat_cols])
    X_test[cat_cols] = cat_imputer.transform(X_test[cat_cols])

scaler = StandardScaler()
X_train.loc[:, num_cols] = scaler.fit_transform(X_train[num_cols])
X_val.loc[:, num_cols] = scaler.transform(X_val[num_cols])
X_test.loc[:, num_cols] = scaler.transform(X_test[num_cols])

# 7. Hyperparameter tuning
alphas = np.logspace(-4, 2, 25)
best_alpha = None
best_r2 = -np.inf
for a in alphas:
    model = Ridge(alpha=a, max_iter=50000)
    model.fit(X_train, y_train)
    r2 = r2_score(y_val, model.predict(X_val))
    if r2 > best_r2:
        best_r2 = r2
        best_alpha = a

print('Best alpha (validation):', best_alpha, 'with R²:', round(best_r2, 4))

# 8. Retrain on train+val
train_val = pd.concat([train, val]).reset_index(drop=True)
y_train_val = train_val['Weekly_Sales']
X_train_val = train_val.drop(columns=['Weekly_Sales'])

for c in original_cat_cols:
    if c + '_freq' not in train_val.columns and c in train_val.columns:
        freq_map = train[c].value_counts(normalize=True).to_dict()
        train_val[c + '_freq'] = train_val[c].map(freq_map).fillna(0)
        train_val = train_val.drop(columns=[c])

X_train_val[num_cols] = num_imputer.transform(X_train_val[num_cols])
X_train_val.loc[:, num_cols] = scaler.transform(X_train_val[num_cols])

model_final = Ridge(alpha=best_alpha, max_iter=50000)
model_final.fit(X_train_val, y_train_val)

# 9. Predictions
y_train_pred = model_final.predict(X_train)
y_val_pred = model_final.predict(X_val)
y_test_pred = model_final.predict(X_test)

# 10. Metrics function
def get_metrics(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    return r2, mse, rmse, mae

# 11. Parity plot function
def parity_plot(y_true, y_pred, title):
    r2, mse, rmse, mae = get_metrics(y_true, y_pred)
    plt.figure(figsize=(6,6))
    plt.scatter(y_true, y_pred, alpha=0.6, edgecolors='k')
    mn, mx = min(y_true.min(), y_pred.min()), max(y_true.max(), y_pred.max())
    plt.plot([mn, mx], [mn, mx], 'r--')
    plt.xlabel('Actual Weekly_Sales')
    plt.ylabel('Predicted Weekly_Sales')
    plt.title(f'Parity Plot: {title}')
    plt.grid(True)
    plt.text(mn, mx, f'R²: {r2:.4f}\nMSE: {mse:.2f}\nRMSE: {rmse:.2f}\nMAE: {mae:.2f}',
             fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.8))
    plt.tight_layout()
    plt.show()

# 12. Generate plots
parity_plot(y_train, y_train_pred, "Training (60%)")
parity_plot(y_val, y_val_pred, "Validation (20%)")
parity_plot(y_test, y_test_pred, "Test (20%)")
```

## 14) Train 25+ Regression Models (including 3 DNNs)

```python
# 14) Train 25+ Regression Models (including 3 DNNs) using existing splits in batches
# Purpose:
#   - Benchmark many algorithms on the same (X_train/X_val/X_test) without re-prep
#   - Run models in batches to avoid kernel OOM
#   - Clear memory between models and DNN runs
#   - Save a results CSV, and visualize the best model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import gc

# Safety: ensure preprocessing cells have been executed
for var_name in ["X_train", "y_train", "X_val", "y_val", "X_test", "y_test"]:
    if var_name not in globals():
        raise RuntimeError(f"Missing {var_name}. Please run the earlier data preparation cells first.")

# === Classical ML imports ===
from sklearn.linear_model import (
    LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge, SGDRegressor,
    PassiveAggressiveRegressor, HuberRegressor, RANSACRegressor, TheilSenRegressor,
    Lars, LassoLars, OrthogonalMatchingPursuit, ARDRegression, QuantileRegressor
)
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor,
    BaggingRegressor, ExtraTreesRegressor, HistGradientBoostingRegressor
)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR, NuSVR, LinearSVR
from sklearn.cross_decomposition import PLSRegression
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.neural_network import MLPRegressor

# === Gradient boosting libraries ===
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

# === Keras (DNNs) ===
# Use Keras with TensorFlow backend
import os
os.environ['KERAS_BACKEND'] = 'tensorflow'
from keras import Input
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import backend as K

# Prefer tuned Ridge alpha from earlier section if available; otherwise fallback to default
try:
    _best_alpha = float(best_alpha)
    ridge_estimator = Ridge(alpha=_best_alpha)
    ridge_name = f"Ridge(alpha={_best_alpha:.4g})"
except Exception:
    ridge_estimator = Ridge()
    ridge_name = "Ridge (default)"

# Define batches to limit memory footprint
batch_light = [
    ("LinearRegression", LinearRegression()),
    (ridge_name, ridge_estimator),
    ("Lasso", Lasso()),
    ("ElasticNet", ElasticNet()),
    ("BayesianRidge", BayesianRidge()),
    ("SGDRegressor", SGDRegressor(max_iter=1000, tol=1e-3)),
    ("PassiveAggressiveRegressor", PassiveAggressiveRegressor(max_iter=1000, tol=1e-3)),
    ("HuberRegressor", HuberRegressor()),
    ("RANSACRegressor", RANSACRegressor()),
    ("TheilSenRegressor", TheilSenRegressor()),
    ("Lars", Lars()),
    ("LassoLars", LassoLars()),
    ("OrthogonalMatchingPursuit", OrthogonalMatchingPursuit()),
    ("ARDRegression", ARDRegression()),
    ("QuantileRegressor", QuantileRegressor()),
    ("DecisionTreeRegressor", DecisionTreeRegressor(random_state=42)),
    ("KNeighborsRegressor", KNeighborsRegressor(n_neighbors=7)),
    ("PLSRegression", PLSRegression(n_components=min(10, X_train.shape[1]))),
]

batch_ensembles = [
    ("RandomForestRegressor", RandomForestRegressor(n_estimators=300, random_state=42, n_jobs=-1)),
    ("ExtraTreesRegressor", ExtraTreesRegressor(n_estimators=300, random_state=42, n_jobs=-1)),
    ("GradientBoostingRegressor", GradientBoostingRegressor(random_state=42)),
    ("HistGradientBoostingRegressor", HistGradientBoostingRegressor(random_state=42)),
    ("AdaBoostRegressor", AdaBoostRegressor(random_state=42)),
    ("BaggingRegressor", BaggingRegressor(random_state=42, n_jobs=-1)),
    ("XGBRegressor", XGBRegressor(n_estimators=400, learning_rate=0.05, subsample=0.9, colsample_bytree=0.9, random_state=42)),
    ("LGBMRegressor", LGBMRegressor(n_estimators=600, learning_rate=0.05, subsample=0.9, colsample_bytree=0.9, random_state=42, verbose=-1)),
    ("CatBoostRegressor", CatBoostRegressor(iterations=600, learning_rate=0.05, depth=8, random_state=42, verbose=0)),
]

batch_kernel_nn = [
    ("SVR", SVR()),
    ("NuSVR", NuSVR()),
    ("LinearSVR", LinearSVR(max_iter=5000, random_state=42)),
    ("GaussianProcessRegressor", GaussianProcessRegressor()),
    ("MLPRegressor", MLPRegressor(max_iter=2000, early_stopping=True, random_state=42)),
]

model_batches = [batch_light, batch_ensembles, batch_kernel_nn]

# results holds tuples: (model_name, test_R2, test_predictions, hyperparams)
results = []

# === Train/evaluate models batch by batch ===
for bi, batch in enumerate(model_batches, start=1):
    print(f"Running batch {bi}/{len(model_batches)} with {len(batch)} models...")
    for name, model in batch:
        try:
            model.fit(X_train, y_train)                     # fit on TRAIN
            y_pred = model.predict(X_test)                  # predict on TEST only
            score = r2_score(y_test, y_pred)
            params = model.get_params() if hasattr(model, 'get_params') else None
            results.append((name, score, y_pred, params))
        except Exception as e:
            print(f"{name} failed: {e}")
            results.append((name, None, None, None))
        finally:
            # Free memory after each model
            del model
            gc.collect()

# === DNN helpers (clear session per model to free memory/VRAM) ===

def run_dnn(name, layer_defs, epochs=15, batch_size=32):
    # Build sequential model with an explicit Input layer
    model = Sequential([Input(shape=(X_train.shape[1],))] + layer_defs)
    model.compile(optimizer='adam', loss='mse')
    # Train with validation for early feedback (we keep epochs modest for speed/memory)
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val), verbose=0)
    # Evaluate on TEST and record R^2
    y_pred = model.predict(X_test, verbose=0).flatten()
    score = r2_score(y_test, y_pred)
    cfg = model.get_config()
    # Important: release graph/session memory after each DNN
    K.clear_session()
    gc.collect()
    return (name, score, y_pred, cfg)

# Three distinct DNN architectures (satisfies 3+ DNN requirement)
results.append(run_dnn("DNN1", [Dense(128, activation='relu'), Dense(64, activation='relu'), Dense(1)]))
results.append(run_dnn("DNN2", [Dense(256, activation='relu'), Dropout(0.2), Dense(128, activation='relu'), Dense(64, activation='relu'), Dense(1)]))
results.append(run_dnn("DNN3", [Dense(512, activation='relu'), Dense(256, activation='relu'), Dense(128, activation='relu'), Dense(1)]))

# === Results summary ===
results_df = pd.DataFrame(results, columns=["Model", "Test_R2", "y_pred", "Hyperparameters"]).sort_values(
    "Test_R2", key=lambda s: s.fillna(-999), ascending=False
).reset_index(drop=True)
print("Top models by R²:\n", results_df[["Model", "Test_R2"]].head(10))

# Scatter plot: Actual vs Predicted for the best model
best_model_name = results_df.loc[0, "Model"]
best_pred = results_df.loc[0, "y_pred"]
plt.figure(figsize=(7, 7))
plt.scatter(y_test, best_pred, alpha=0.5, edgecolors='k')
mn, mx = min(y_test.min(), np.min(best_pred)), max(y_test.max(), np.max(best_pred))
plt.plot([mn, mx], [mn, mx], 'r--')
plt.xlabel('Actual Weekly_Sales')
plt.ylabel('Predicted Weekly_Sales')
plt.title(f'Actual vs Predicted: {best_model_name}')
plt.grid(True)
plt.show()

# Bar chart of top 10 models by R²
top10 = results_df.dropna(subset=["Test_R2"]).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10["Model"], top10["Test_R2"], color='teal')
plt.xlabel('Test R²')
plt.title('Top 10 Models by Test R²')
plt.gca().invert_yaxis()
plt.grid(True)
plt.tight_layout()
plt.show()

# Save results without bulky predictions array
results_out = results_df.drop(columns=["y_pred"]).copy()
results_out.to_csv('model_comparison_results.csv', index=False)
results_out.head(12)
```

```python
# 14b) Compact hyperparameters table for top-N models
import pandas as pd

if 'results_df' not in globals():
    raise RuntimeError("results_df not found. Run the 25+ models cell first.")

TOP_N = 29

keys_of_interest = [
    'alpha', 'n_estimators', 'max_depth', 'learning_rate', 'subsample',
    'colsample_bytree', 'depth', 'kernel', 'C', 'epsilon', 'n_neighbors',
    'n_components', 'random_state'
]

def summarize_hparams(hparams):
    if hparams is None:
        return {"summary": None}
    # sklearn/xgb/lgbm/catboost dict of params
    if isinstance(hparams, dict):
        out = {k: hparams.get(k) for k in keys_of_interest if k in hparams}
        if len(out) == 0:
            # fallback small subset to avoid huge prints
            for k in list(hparams.keys())[:6]:
                out[k] = hparams[k]
        return out
    # DNN config (keras get_config)
    try:
        layers = hparams.get('layers', []) if isinstance(hparams, dict) else []
        return {"layers_count": len(layers)}
    except Exception:
        return {"summary": "see config"}

compact_rows = []
for _, row in results_df.dropna(subset=['Test_R2']).head(TOP_N).iterrows():
    summary = summarize_hparams(row['Hyperparameters'])
    compact_rows.append({"Model": row['Model'], "Test_R2": row['Test_R2'], **summary})

hyperparams_df = pd.DataFrame(compact_rows)
print(hyperparams_df)
hyperparams_df.to_csv('top_model_hyperparameters.csv', index=False)
hyperparams_df
```

```python
# Generating Ridge R² line plot across alpha values for training and validation sets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

# Load data
df = pd.read_csv('walmart-sales-dataset-of-45stores.csv', parse_dates=['Date'], dayfirst=True)
df = df.dropna(subset=['Weekly_Sales']).copy()

# Feature engineering
df = df.sort_values('Date').reset_index(drop=True)
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.dayofweek
group_cols = [c for c in ['Store','Dept'] if c in df.columns]
if group_cols:
    df = df.sort_values(group_cols + ['Date']).reset_index(drop=True)
    df['Sales_Lag1'] = df.groupby(group_cols)['Weekly_Sales'].shift(1)
    df['Sales_Roll4'] = df.groupby(group_cols)['Weekly_Sales'].transform(lambda x: x.shift(1).rolling(4).mean())
else:
    df['Sales_Lag1'] = df['Weekly_Sales'].shift(1)
    df['Sales_Roll4'] = df['Weekly_Sales'].shift(1).rolling(4).mean()
df = df.dropna().reset_index(drop=True)
df = df.drop(columns=['Date'])

# Frequency encode Store/Dept
for c in ['Store','Dept']:
    if c in df.columns:
        freq_map = df[c].value_counts(normalize=True).to_dict()
        df[c + '_freq'] = df[c].map(freq_map).fillna(0)
        df.drop(columns=[c], inplace=True)

# Time-aware split
n = len(df)
cut1 = int(n * 0.6)
cut2 = int(n * 0.8)
train = df.iloc[:cut1].copy()
val = df.iloc[cut1:cut2].copy()
test = df.iloc[cut2:].copy()

# Define X/y
y_train = train['Weekly_Sales']
X_train = train.drop(columns=['Weekly_Sales'])
y_val = val['Weekly_Sales']
X_val = val.drop(columns=['Weekly_Sales'])

# Impute and scale
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
num_imputer = SimpleImputer(strategy='mean')
X_train[num_cols] = num_imputer.fit_transform(X_train[num_cols])
X_val[num_cols] = num_imputer.transform(X_val[num_cols])
scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_val[num_cols] = scaler.transform(X_val[num_cols])

# Ridge tuning
alphas = np.logspace(-4, 2, 25)
train_r2 = []
val_r2 = []
best_alpha = None
best_val_r2 = -np.inf
for a in alphas:
    model = Ridge(alpha=a, max_iter=10000)
    model.fit(X_train, y_train)
    train_r2.append(r2_score(y_train, model.predict(X_train)))
    val_score = r2_score(y_val, model.predict(X_val))
    val_r2.append(val_score)
    if val_score > best_val_r2:
        best_val_r2 = val_score
        best_alpha = a

# Plot
plt.figure(figsize=(8,5))
plt.plot(alphas, train_r2, label='Train R²', marker='o')
plt.plot(alphas, val_r2, label='Validation R²', marker='s')
plt.scatter(best_alpha, best_val_r2, color='red', zorder=5, label=f'Best alpha: {best_alpha:.4f}')
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('R² Score')
plt.title('Ridge Regression R² Scores Across Alpha Values')
plt.grid(True)
plt.legend()
plt.tight_layout()
#plt.savefig('ridge_r2_alpha_plot.png')
print("Plot saved as ridge_r2_alpha_plot.png")

```
