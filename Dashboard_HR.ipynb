```python
import streamlit as st
import pandas as pd
import plotly.express as px

# Path to your processed Excel file
file_path = "/Users/jaacabrera/Documents/Python Scripts/data_io/Resigned_Tracker_2019_2025.xlsx"

# Load sheets
resigned_df = pd.read_excel(file_path, sheet_name="Resigned Employees")
joined_df = pd.read_excel(file_path, sheet_name="Joined Employees")
summary_resigned = pd.read_excel(file_path, sheet_name="Resignation Summary")
summary_joined = pd.read_excel(file_path, sheet_name="Join Summary")
growth_df = pd.read_excel(file_path, sheet_name="Resource Growth Tracker")

# --- Dashboard Layout ---
st.set_page_config(page_title="HR Analytics Dashboard", layout="wide")
st.title("📊 HR Analytics Dashboard (2019–2025)")

# --- Filters ---
years = sorted(resigned_df["Resigned Year"].dropna().unique())
selected_year = st.selectbox("Select Year", years)

months = list(resigned_df["Resigned Month Name"].unique())
selected_months = st.multiselect("Select Months", months, default=months)

# --- Filtered Data ---
filtered_resigned = resigned_df[
    (resigned_df["Resigned Year"] == selected_year) &
    (resigned_df["Resigned Month Name"].isin(selected_months))
]

filtered_joined = joined_df[
    (joined_df["Joined Year"] == selected_year) &
    (joined_df["Joined Month Name"].isin(selected_months))
]

# --- Resignation Trends ---
st.header("Resignation Trends")
fig_resigned = px.bar(
    filtered_resigned,
    x="Resigned Month Name",
    color="Resigned Year",
    title=f"Resignations in {selected_year}",
)
st.plotly_chart(fig_resigned, use_container_width=True)

# --- Join Trends ---
st.header("Join Trends")
fig_joined = px.bar(
    filtered_joined,
    x="Joined Month Name",
    color="Joined Year",
    title=f"Joins in {selected_year}",
)
st.plotly_chart(fig_joined, use_container_width=True)

# --- Resource Growth ---
st.header("Resource Growth Tracker")
fig_growth = px.line(
    growth_df,
    x="Year",
    y="Ending Headcount",
    title="Headcount Growth Over Time",
    markers=True
)
st.plotly_chart(fig_growth, use_container_width=True)

# --- Data Tables ---
st.header("Detailed Data")
st.subheader("Resigned Employees (Filtered)")
st.dataframe(filtered_resigned)

st.subheader("Joined Employees (Filtered)")
st.dataframe(filtered_joined)

st.subheader("Resource Growth Tracker")
st.dataframe(growth_df)
```
